# Zomato Data analysis project

# Step 1 - Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - Import the Data

```python
dataframe = pd.read_csv("Zomato_data .csv")
```

```python
dataframe
```

# Step 3 - Cleaining the data

```python
def extract_rating_number(value):
    value = value.split("/")
    value = value[0]
    return float(value)
```

```python
dataframe["rate"] = dataframe["rate"].apply(extract_rating_number)
```

```python
dataframe
```

```python
dataframe.info()
```

# Type of restaurent do the majority of customers order from 

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe["listed_in(type)"])
plt.xlabel("Type of restaurent")
plt.title("Majority of customers order from")
plt.show()
```

# Votes each type of restaurent recieved

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({ "votes":grouped_data })
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurent",)
plt.ylabel("Votes")
plt.title("Votes each type of restaurent recieved",fontweight="bold")
plt.show()
```

# Ratings that majority of reastaurents have 

```python
dataframe.head()
```

```python
plt.hist(dataframe["rate"],bins=5)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Ratings that majority of restaurents have",fontweight="bold")
plt.show()
```

# Average spending couple on each other

```python
dataframe.head()
```

```python
couple_data = dataframe["approx_cost(for two people)"]
sns.countplot(x=couple_data)
plt.xlabel("Average spending")
plt.ylabel("People")
plt.title("Average spending couple on each other",fontweight="bold")
plt.show()
```

# Which mode recieve maximum rating

```python
dataframe.head()
```

```python
sns.boxplot(x="online_order",y="rate",data=dataframe)
plt.xlabel("Online order")
plt.ylabel("Rate")
plt.title("Which mode recieve maximum rating",fontweight="bold")
plt.show()
```

# Type of students recieved more offline orders

```python
pivot_table = dataframe.pivot_table(index="listed_in(type)",columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.title("Type of students recieved more offline orders",fontweight="bold")
plt.show()
```
